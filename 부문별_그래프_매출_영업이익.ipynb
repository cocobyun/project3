```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# NanumGothic 폰트 설정
!apt-get -qq install fonts-nanum
font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"
fontprop = fm.FontProperties(fname=font_path, size=12)

# 데이터 준비
sales_data = pd.DataFrame({
    '부문': ['택배·이커머스', '계약물류(CL)', '글로벌'],
    '매출액(억원)': [8982, 7684, 10985],
    '영업이익(억원)': [540, 512, 282],
    '매출 YoY(%)': [0, 6.2, 4.2],
    '영업이익 YoY(%)': [-3.1, 15.8, 83.1]
})

# 막대의 너비와 x축 위치 조정
bar_width = 0.35
index = np.arange(len(sales_data['부문']))

# 보조축을 추가하여 매출액과 영업이익을 같은 그래프에 표시
fig, ax1 = plt.subplots(figsize=(14, 8))

# 첫 번째 축: 매출액
bars1 = ax1.bar(index, sales_data['매출액(억원)'], bar_width, label='매출액(억원)', color='skyblue')
ax1.set_xlabel('부문', fontsize=14, fontproperties=fontprop)
ax1.set_ylabel('매출액(억원)', fontsize=14, color='skyblue', fontproperties=fontprop)
ax1.tick_params(axis='y')
ax1.set_xticks(index + bar_width / 2)
ax1.set_xticklabels(sales_data['부문'], fontsize=12, fontproperties=fontprop)

# 주석 추가 (매출 YoY)
for bar, yoy in zip(bars1, sales_data['매출 YoY(%)']):
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval , f'{yoy}%', va='bottom', ha='center', fontsize=12)

# 두 번째 축: 영업이익
ax2 = ax1.twinx()  # 보조축 생성
bars2 = ax2.bar(index + bar_width, sales_data['영업이익(억원)'], bar_width, label='영업이익(억원)', color='lightgreen')
ax2.set_ylabel('영업이익(억원)', fontsize=14, color='lightgreen', fontproperties=fontprop)
ax2.tick_params(axis='y')

# 주석 추가 (영업이익 YoY)
for bar, yoy in zip(bars2, sales_data['영업이익 YoY(%)']):
    yval = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2, yval, f'{yoy}%', va='bottom', ha='center', fontsize=12)

# 제목 설정
plt.title('3Q 매출액 및 영업이익', fontsize=16, pad=20, fontproperties=fontprop)

ax1.set_ylim(0, 13000)  # 매출액 축의 최댓값 설정
ax2.set_ylim(0, 1000)   # 영업이익 축의 최댓값 설정

plt.show()
```
