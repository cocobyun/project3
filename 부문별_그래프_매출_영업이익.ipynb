```python
#Load the data into a new DataFrame for visualization

# NanumGothic  
!apt-get -qq install fonts-nanum  # Nanum  
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

#  
font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"
fontprop = fm.FontProperties(fname=font_path, size=12)


naver_sales_data = pd.DataFrame({
    'CJ 3Q ': ['', '(CL)', ''],
    '3Q  ()': [9314, 2569, 1562, 2070, 1475],
    'YoY (%)': [-6.5, -1.4, -18.3, 6.8, 3.6]
})

# Create a bar chart for 3Q sales with adjusted bar width
plt.figure(figsize=(14, 8))  #     
bar_width = 0.5  #   
bars = plt.bar(naver_sales_data['KT 3Q '], naver_sales_data['3Q  ()'], color='lightpink', width=bar_width)

# Annotate the YoY growth percentage next to each bar
for bar, yoy in zip(bars, naver_sales_data['YoY (%)']):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval+200, f'{yoy}%', va='bottom', ha='center', fontsize=14)  #   

# Adjust labels and title with increased fontsize and custom spacing
plt.xlabel('', fontproperties=fontprop, fontsize=16, labelpad=20)  # x     
plt.ylabel('3Q  ()', fontproperties=fontprop, fontsize=16, labelpad=15)
plt.title('KT 3Q ', fontproperties=fontprop, fontsize=20, pad=30)  #     

# Set y-axis limit
plt.ylim(0, 10500)  # y  12000 
plt.gca().xaxis.set_tick_params(pad=10)
plt.tight_layout()

#  
plt.xticks(fontproperties=fontprop, fontsize=14)  # x    
plt.yticks(fontproperties=fontprop, fontsize=14)  # y    
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# NanumGothic  
!apt-get -qq install fonts-nanum
font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"
fontprop = fm.FontProperties(fname=font_path, size=12)

#  
sales_data = pd.DataFrame({
    '': ['', '(CL)', ''],
    '()': [8982, 7684, 10985],
    '()': [540, 512, 282],
    ' YoY(%)': [0, 6.2, 4.2],
    ' YoY(%)': [-3.1, 15.8, 83.1]
})

#   x  
bar_width = 0.35
index = np.arange(len(sales_data['']))

#       
fig, ax1 = plt.subplots(figsize=(14, 8))

#   : 
bars1 = ax1.bar(index, sales_data['()'], bar_width, label='()', color='skyblue')
ax1.set_xlabel('', fontsize=14, fontproperties=fontprop)
ax1.set_ylabel('()', fontsize=14, color='skyblue', fontproperties=fontprop)
ax1.tick_params(axis='y')
ax1.set_xticks(index + bar_width / 2)
ax1.set_xticklabels(sales_data[''], fontsize=12, fontproperties=fontprop)

#   ( YoY)
for bar, yoy in zip(bars1, sales_data[' YoY(%)']):
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval , f'{yoy}%', va='bottom', ha='center', fontsize=12)

#   : 
ax2 = ax1.twinx()  #  
bars2 = ax2.bar(index + bar_width, sales_data['()'], bar_width, label='()', color='lightgreen')
ax2.set_ylabel('()', fontsize=14, color='lightgreen', fontproperties=fontprop)
ax2.tick_params(axis='y')

#   ( YoY)
for bar, yoy in zip(bars2, sales_data[' YoY(%)']):
    yval = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2, yval, f'{yoy}%', va='bottom', ha='center', fontsize=12)

#  
plt.title('3Q   ', fontsize=16, pad=20, fontproperties=fontprop)

ax1.set_ylim(0, 13000)  #    
ax2.set_ylim(0, 1000)   #    

plt.show()
```
