```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load the data into a new DataFrame for visualization

# NanumGothic 폰트 설정
!apt-get -qq install fonts-nanum  # Nanum 폰트 설치
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 폰트 설정
font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"
fontprop = fm.FontProperties(fname=font_path, size=12)


naver_sales_data = pd.DataFrame({
    'KT 3Q 매출': ['BC카드', '스카이라이프', '콘텐츠자회사', 'kt cloud', '에스테이트'],
    '3Q 매출 (억원)': [9314, 2569, 1562, 2070, 1475],
    'YoY (%)': [-6.5, -1.4, -18.3, 6.8, 3.6]
})

# Create a bar chart for 3Q sales with adjusted bar width
plt.figure(figsize=(14, 8))  # 차트 크기를 더 크게 조정
bar_width = 0.5  # 막대의 두께 설정
bars = plt.bar(naver_sales_data['KT 3Q 매출'], naver_sales_data['3Q 매출 (억원)'], color='lightpink', width=bar_width)

# Annotate the YoY growth percentage next to each bar
for bar, yoy in zip(bars, naver_sales_data['YoY (%)']):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval+200, f'{yoy}%', va='bottom', ha='center', fontsize=14)  # 글씨 크기 조정

# Adjust labels and title with increased fontsize and custom spacing
plt.xlabel('부문별', fontproperties=fontprop, fontsize=16, labelpad=20)  # x축 타이틀과 그래프 간의 간격 조정
plt.ylabel('3Q 매출 (억원)', fontproperties=fontprop, fontsize=16, labelpad=15)
plt.title('KT 3Q 매출', fontproperties=fontprop, fontsize=20, pad=30)  # 제목과 차트 간의 간격 조정

# Set y-axis limit
plt.ylim(0, 10500)  # y축의 최대값을 12000으로 설정
plt.gca().xaxis.set_tick_params(pad=10)
plt.tight_layout()

# 그래프 표시
plt.xticks(fontproperties=fontprop, fontsize=14)  # x축 눈금 글씨 크기 조정
plt.yticks(fontproperties=fontprop, fontsize=14)  # y축 눈금 글씨 크기 조정
plt.show()
```
