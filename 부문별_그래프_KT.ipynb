```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load the data into a new DataFrame for visualization

# NanumGothic  
!apt-get -qq install fonts-nanum  # Nanum  
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

#  
font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"
fontprop = fm.FontProperties(fname=font_path, size=12)


naver_sales_data = pd.DataFrame({
    'KT 3Q ': ['BC', '', '', 'kt cloud', ''],
    '3Q  ()': [9314, 2569, 1562, 2070, 1475],
    'YoY (%)': [-6.5, -1.4, -18.3, 6.8, 3.6]
})

# Create a bar chart for 3Q sales with adjusted bar width
plt.figure(figsize=(14, 8))  #     
bar_width = 0.5  #   
bars = plt.bar(naver_sales_data['KT 3Q '], naver_sales_data['3Q  ()'], color='lightpink', width=bar_width)

# Annotate the YoY growth percentage next to each bar
for bar, yoy in zip(bars, naver_sales_data['YoY (%)']):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval+200, f'{yoy}%', va='bottom', ha='center', fontsize=14)  #   

# Adjust labels and title with increased fontsize and custom spacing
plt.xlabel('', fontproperties=fontprop, fontsize=16, labelpad=20)  # x     
plt.ylabel('3Q  ()', fontproperties=fontprop, fontsize=16, labelpad=15)
plt.title('KT 3Q ', fontproperties=fontprop, fontsize=20, pad=30)  #     

# Set y-axis limit
plt.ylim(0, 10500)  # y  12000 
plt.gca().xaxis.set_tick_params(pad=10)
plt.tight_layout()

#  
plt.xticks(fontproperties=fontprop, fontsize=14)  # x    
plt.yticks(fontproperties=fontprop, fontsize=14)  # y    
plt.show()
```
