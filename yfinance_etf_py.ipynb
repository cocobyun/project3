```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# NanumGothic 폰트 설정
# !apt-get -qq install fonts-nanum  # Uncomment this line if you're running in a Jupyter environment where you can install packages

# 폰트 설정
font_path = "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"
fontprop = fm.FontProperties(fname=font_path, size=12)

tickers = {
    '자동차': '305720.KS',
    '반도체': '404960.KS',
    '은행': '091170.KS',
    '헬스케어': '235980.KS',
    '철강': '266150.KS',
    '건설': '117700.KS',
    '증권': '102970.KS',
    'IT': '251370.KS',
    '200IT TR': '217780.KS',
    '미디어&엔터테인먼트': '234300.KS',
    '에너지화학': '117460.KS',
    '운송': '333620.KS',
    '기계장비': '304700.KS',
    '필수소비재': '253220.KS',
    '경기소비재': '279600.KS',
    '보험': '140700.KS',
}

# 데이터 불러오기 (10/28일부터 11/1일까지)
start_date = '2024-10-28'
end_date = '2024-11-01'
# 빈 데이터프레임 생성
df = pd.DataFrame()

# 각 티커에 대해 주식 데이터 다운로드
for sector, ticker in tickers.items():
    # 'Adj Close'로 종가 불러오기
    data = yf.download(ticker, start=start_date, end=end_date)['Adj Close']
    df[sector] = data

# 주간 수익률 계산
returns = (df.iloc[-1] - df.iloc[0]) / df.iloc[0] * 100

plt.figure(figsize=(10, 6))
plt.bar(returns.index, returns.values, color=['green' if r > 0 else 'red' for r in returns])

# 그래프 설정
plt.title('10월 5주차 수익률', fontproperties=fontprop)  # 제목에 폰트 적용
plt.xlabel('sector(kodex etf)', fontproperties=fontprop)  # x축 레이블에 폰트 적용
plt.ylabel('returns (%)', fontproperties=fontprop)  # y축 레이블에 폰트 적용

# Rotate x-axis labels to be vertical
plt.xticks(rotation=90, fontproperties=fontprop)  # x축 눈금에 폰트 적용 및 회전
plt.yticks(fontproperties=fontprop)  # y축 눈금에 폰트 적용

# Set y-axis limits to ensure visibility of all values
plt.ylim(min(returns.values) - 2, max(returns.values) + 2)  # Adjust the limits as needed

# Optionally, set y-ticks to clearly show negative values
plt.yticks(fontproperties=fontprop)  # y축 눈금에 폰트 적용

# 그래프 표시
plt.tight_layout()
plt.show()
```
